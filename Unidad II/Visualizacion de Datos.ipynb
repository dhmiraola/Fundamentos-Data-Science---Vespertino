# Visualizacion de datos usando Matplotlib

## Histogramas
* Los histogramas son útiles para visualizar la distribución de una variable numérica. Permiten detectar cualquier anomalía como valores atipicos.

```python
plt.subplots(figsize=(7,4))
plt.hist(data['Unit price'])
plt.title('Distribución del Precio Unitario')
plt.xlabel('Precio Unitario')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.subplots(figsize=(7,4))
plt.hist(data['Quantity'])
plt.title('Distribución de la Cantidad')
plt.xlabel('Cantidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.subplots(figsize=(7,4))
plt.hist(data['gross income'])
plt.title('Distribución del Ingreso Bruto')
plt.xlabel('Ingreso Bruto')
plt.ylabel('Frecuencia')
plt.show()
```

## Grafico de Dispersion
* Son excelentes para visualizar la relación entre dos variables numéricas, permitiéndote identificar patrones, tendencias y correlaciones potenciales entre ellas.

```python
plt.subplots(figsize=(7,4))
plt.scatter(x=data["Total"],y=data["gross income"])
plt.xlabel("Precio Total")
plt.ylabel("Ingresos Brutos")
plt.title("Relación entre el Precio Total y los Ingresos Brutos")
plt.show()
```

```python
plt.subplots(figsize=(7,4))
plt.scatter(x=data["Unit price"],y=data["gross income"])
plt.xlabel("Precio Unitario")
plt.ylabel("Ingresos Brutos")
plt.title("Relación entre la el Precio Unitario y los Ingresos Brutos")
plt.show()
```

```python
plt.subplots(figsize=(7,4))
plt.scatter(x=data["Quantity"],y=data["cogs"])
plt.xlabel("Numero de Productos Adquiridos por el Cliente")
plt.ylabel("Costo de los Bienes Vendidos")
plt.title("Relación entre El numero de Productos Adquiridos por el Cliente y El Costo de los Bienes")
plt.show()
```

## Grafico de Torta o Pie
* Son ideales para mostrar proporciones o porcentajes entre categorías. Son más efectivos cuando se tienen pocas categorías y se desea destacar la contribución de cada categoría al total.

```python
genero = data['Gender'].value_counts()
plt.pie(genero, labels=genero.index, autopct='%1.1f%%')
plt.title('Distribución de Género')
plt.show()
```

```python
productos = data['Product line'].value_counts()
plt.pie(productos, labels=productos.index, autopct='%1.1f%%')
plt.title('Distribución de la Linea de Productos')
plt.show()
```

```python
productos = data['Payment'].value_counts()
plt.pie(productos, labels=productos.index, autopct='%1.1f%%')
plt.title('Distribución de la Forma de Pago')
plt.show()
```

## Grafico de Barras
* Los gráficos de barras son excelentes para comparar cantidades entre categorías. Son útiles para visualizar diferencias entre grupos, como comparar el conteo de elementos en categorías discretas.

```python
ciudad = data['City'].value_counts()
plt.bar(ciudad.index, ciudad.values)
plt.title('Vehículos Eléctricos por Uso Principal')
plt.xlabel('Uso Principal')
plt.xticks(rotation=45)
plt.ylabel('Cantidad')
plt.show()
```

```python
ciudad = data['Product line'].value_counts()
plt.bar(ciudad.index, ciudad.values)
plt.title('Lineas de Productos')
plt.xlabel('Tipos de Productos')
plt.xticks(rotation=75)
plt.ylabel('Cantidad')
plt.show()
```

```python
pagos = data['Payment'].value_counts()
plt.bar(pagos.index, pagos.values)
plt.title('Formas de pago')
plt.xlabel('Formasde Pago')
plt.xticks(rotation=75)
plt.ylabel('Cantidad')
plt.show()
```

# Visualizacion de datos utilizando Seaborn

```python
# Primero debemos importar la libreria seaborn 

import seaborn as sns
```

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=data, x="Rating", hue="Product line", multiple="stack", palette="viridis")
plt.title("Distribución de Calificaciones por Línea de Producto")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.legend(title='Línea de Producto')
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=data, x="Payment", palette="autumn")
plt.title("Distribución de Métodos de Pago Utilizados")
plt.xlabel("Método de Pago")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x=data["City"], y = data["Total"], hue =  data["Gender"], palette="autumn")
plt.title("Balance según país y género")
plt.ylabel("Balance")
plt.xlabel("País")
plt.show()
```
